```python
# -------------------------------
# STEP 1: Upload File (Choose Files)
# -------------------------------
from google.colab import files
import pandas as pd
import numpy as np
import zipfile
import os

uploaded = files.upload()  # choose your CSV (or ZIP)

file_name = list(uploaded.keys())[0]

# -------------------------------
# STEP 2: Handle ZIP or CSV
# -------------------------------
if file_name.endswith('.zip'):
    with zipfile.ZipFile(file_name, 'r') as zip_ref:
        zip_ref.extractall('/content')
    print("✅ ZIP Extracted")

    # find first CSV inside
    csv_files = []
    for root, dirs, files_list in os.walk('/content'):
        for f in files_list:
            if f.endswith('.csv'):
                csv_files.append(os.path.join(root, f))

    if len(csv_files) == 0:
        raise Exception("❌ No CSV found inside ZIP")

    file_path = csv_files[0]
else:
    file_path = file_name

print("📂 Using file:", file_path)

# -------------------------------
# STEP 3: Load Dataset
# -------------------------------
data = pd.read_csv(file_path)

print("\nDataset Loaded!\n")
print(data.head())

# -------------------------------
# STEP 4: Preprocessing
# -------------------------------
# Convert categorical → numeric (if any)
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for col in data.columns:
    if data[col].dtype == 'object':
        data[col] = le.fit_transform(data[col])

# -------------------------------
# STEP 5: Linear Regression (1 feature)
# -------------------------------
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

print("\n--- Linear Regression (Single Feature) ---")

X = data.iloc[:, [0]]   # first column as feature
y = data.iloc[:, -1]    # last column as target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model_lr = LinearRegression()
model_lr.fit(X_train, y_train)

y_pred = model_lr.predict(X_test)

print("MSE:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))

# -------------------------------
# STEP 6: Multiple Linear Regression
# -------------------------------
print("\n--- Multiple Linear Regression ---")

X_multi = data.iloc[:, :-1]   # all features
y_multi = data.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X_multi, y_multi, test_size=0.2)

model_mlr = LinearRegression()
model_mlr.fit(X_train, y_train)

y_pred = model_mlr.predict(X_test)

print("MSE:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))

# -------------------------------
# STEP 7: New Sample Prediction
# -------------------------------
print("\nEnter new sample (comma-separated values):")
print("Example:", ",".join(["0"] * X_multi.shape[1]))

sample = list(map(float, input("Input: ").split(",")))
prediction = model_mlr.predict([sample])

print("Predicted Value:", prediction)
```
